we see that for running cycling unocycling the avg heart rate are differen so we can replace the nan values by avg of each type

```python
df[df['Type'] == 'Running']['Average Heart Rate (bpm)'].mean()
```

```python
#craeting x as a df to replace the na with avg heart rate of each type 
x=df.copy()
```

x['Type'] == t: This part of the code creates a boolean mask that returns True for rows where the 'Type' column of DataFrame 'x' matches the value of 't'.
x['Average Heart Rate (bpm)'].isna(): This part of the code creates another boolean mask that returns True for rows where the 'Average Heart Rate (bpm)' column of DataFrame 'x' has missing or NaN values.
(x['Type'] == t) & (x['Average Heart Rate (bpm)'].isna()): This part combines the two boolean masks using the '&' operator to return True only for rows where both conditions are satisfied.
x.loc[boolean_mask, 'Average Heart Rate (bpm)'] = m: This part uses the .loc accessor of the DataFrame object to select only the rows where the boolean mask is True and then sets the value of the 'Average Heart Rate (bpm)' column of those rows to the value of 'm'.

```python

# group the DataFrame by type and compute the mean for each type
mean_rates = x.groupby('Type')['Average Heart Rate (bpm)'].mean()

# loop over each type and replace NaN values in the 'rate' column with the mean value for that type
for t, m in mean_rates.iteritems():
    x.loc[(x['Type'] == t) & (x['Average Heart Rate (bpm)'].isna()), 'Average Heart Rate (bpm)'] = m

# print the updated DataFrame
x
```

```python
# creting different data frame based on type and also  to replce nan values 

mean_run=df[df['Type']=='Running']['Average Heart Rate (bpm)'].mean()
mean_cycle=df[df['Type']=='Cycling']['Average Heart Rate (bpm)'].mean()


df_run=df[df['Type']=="Running"].copy()
df_walk=df[df['Type']=='Walking'].copy()
df_cyc=df[df['Type']=="Cycling"].copy()
```

Pandas dataframe.resample() function is primarily used for time series data.
A time series is a series of data points indexed (or listed or graphed) in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. It is a Convenience method for frequency conversion and resampling of time series. Object must have a datetime-like index (DatetimeIndex, PeriodIndex, or TimedeltaIndex), or pass datetime-like values to the on or level keyword.

```python
import statsmodels.api as sm

df_week_dist=df_run['2013':'2018']['Distance (km)'].resample('W').bfill()
decompose=sm.tsa.seasonal_decompose(df_week_dist,extrapolate_trend=1,period=52)

fig=plt.figure(figsize=(12,5))

ax=decompose.trend.plot(label='Trend',linewidth=2)
ax=decompose.observed.plot(label='Observed',linewidth=0.5)

ax.legend()
ax.set_title('running distance trend ')
plt.show()

```
